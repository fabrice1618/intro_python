```python
import pandas as pd

excel_file = 'coupe_monde.xlsx'

df = pd.read_excel(excel_file, usecols=list(range(1,3)) )

print("Accès aux colonnes par le nom:")
for i in range(len(df)): 
    print(df.loc[i,"Année"], df.loc[i,"Vainqueur"] )
    
print("\nAccès aux colonnes par l'index:")    
for i in range(5, 10): 
    print(df.iloc[i,0], df.iloc[i,1] )
```

```python
# itération par ligne

victoires = {}

for index, row in df.iterrows(): 
    print(index, row["Année"], row["Vainqueur"])
    
    vainqueur = row["Vainqueur"]
    if vainqueur in victoires:
        victoires[vainqueur][0] += 1
    else:
        victoires[vainqueur] = [1]

print(victoires)
```

```python
#victoires.to_excel(writer, index=False, sheet_name='report')
victoire_dataframe = pd.DataFrame.from_dict(victoires)
victoire_dataframe.to_excel('victoires.xlsx')
```

```python
import csv

def transforme_pays(pays):
    if pays == "Allemagne de l'Ouest":
        return "Allemagne"
    else:
        return pays



# ouverture fichier CSV
nombre_ligne = 0
with open('coupe_monde.csv', encoding="utf8") as fichier_csv:
    lecteur_csv = csv.reader(fichier_csv, delimiter=';')
    
    victoires = {}
    
    for ligne in lecteur_csv:
        if nombre_ligne > 0:
            # c'est une ligne de données
            #print(ligne)
            pays = transforme_pays( ligne[2] )
            pays2 = transforme_pays( ligne[4] )
            pays3 = transforme_pays( ligne[5] )

            # Ajouter une victoire
            if pays in victoires:
                # dejà présent, ajouter une victoire
                victoires[pays][0] += 1
            else:
                # le pays n'est pas présent
                victoires[pays] = [1, 0, 0]
                
            #ajouter 2eme place
            if pays2 in victoires:
                # dejà présent, ajouter une victoire
                victoires[pays2][1] += 1
            else:
                # le pays n'est pas présent
                victoires[pays2] = [0, 1, 0]
                
            #ajouter 3eme place
            if pays3 in victoires:
                # dejà présent, ajouter une victoire
                victoires[pays3][2] += 1
            else:
                # le pays n'est pas présent
                victoires[pays3] = [0, 0, 1]
        #else:
            # Afficher le titre
            #print("Titre")
            #print(ligne)
            #print()
        nombre_ligne += 1
        
    print(victoires)
        
```

```python
victoire_dataframe = pd.DataFrame.from_dict(victoires)
victoire_dataframe.to_excel('victoires.xlsx')
```
