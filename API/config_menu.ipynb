```python
from ConfigJSON import ConfigJSON

config = ConfigJSON("config.json")

def print_menu():
    print("Menu:")
    print("E: exist_file")
    print("P: print")
    print("G: get_config")
    print("S: set_config")
    print("D: delete_config")
    print("R: read_file")
    print("W: write_file")
    print("Q: Quitter\n")


choix = ""

print_menu()

while choix.upper() != "Q":
    
    choix = input("Votre choix > ")
    choix = choix.upper()

    if choix == "P":
        print(config)

    if choix == "H":
        print_menu()
        
    if choix == "E":
        if config.exist_file():
            print(f"fichier {config.file_name()} existant")
        else:
            print(f"fichier {config.file_name()} inexistant")
        
    if choix == "G":
        param = input("param: ")
        print( f"get_config( '{param}' ): '{config.get_config(param)}'")

    if choix == "S":
        param = input("param: ")
        valeur = input("valeur: ")
        config.set_config(param, valeur)
        print( f"set_config( '{param}', '{valeur}' )")

    if choix == "D":
        param = input("param: ")
        config.delete_config(param)
        print( f"delete_config( '{param}' )")

    if choix == "R":
        config.read_file()
        print( "read_file()")

    if choix == "W":
        config.write_file()
        print( "write_file()")
        
print("Terminé")
```
